Prepareer je data in je Jupyter Notebook:
One-hot encode je uitgekozen categorische feature(s)

```python
df = df.loc[:, ['SalePrice', 'Gr Liv Area', 'Neighborhood', 'Overall Qual', 'Lot Area', 'House Style']]
# 
df
```

```python
dummies_dataframe = pd.get_dummies(df.loc[:, ['Neighborhood' , 'House Style']], drop_first=True)
#
dummies_dataframe
```

```python
df = pd.concat([df, dummies_dataframe], axis=1)
df = df.drop(['Neighborhood', 'House Style'], axis=1)
#
df
```

Splits de dataset horizontaal én verticaal, zodat er vier “stukken” ontstaan.

```python
X = df.loc[df['SalePrice'].notna(), :]
X = X.drop('SalePrice', axis=1)
Y = df.loc[df['SalePrice'].notna(), 'SalePrice']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size= 0.15,
    random_state=1
)
```

Initialiseer een lineair regressiemodel in je Jupyter Notebook en train deze op de juiste “stukken” uit stap 4b. Doe hierbij onderzoek naar de zogeheten “hyperparameters” die je bij dit model kan instellen met behulp van de pythonfunctie help(). Kies hierbij voor een willekeurige combinatie van minstens 2 hyperparameters. Noteer deze laatste twee zgn. “hyperparameters” in je logboek.

```python
reg_model = LinearRegression()
reg_model.fit(X_train, Y_train)
y_pred = reg_model.predict(X_test)
```

```python
print(mean_squared_error(Y_test, y_pred))
print(mean_absolute_error(Y_test, y_pred))
print(reg_model.score(X_test, Y_test))
```
