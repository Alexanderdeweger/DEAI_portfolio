Prepareer je data in je Jupyter Notebook:
    -One-hot encode je uitgekozen categorische feature(s)
    -Splits de dataset horizontaal én verticaal, zodat er vier “stukken” ontstaan.

```python
#dataset = dataset.loc[:,['SalePrice', 'Overall Qual',	'Gr Liv Area',	'Neighborhood', 'House Style']]
dataset
```

```python
Style1 = pd.get_dummies(dataset.loc[:, 'Neighborhood'], drop_first=True)
Style1
Style2 = pd.get_dummies(dataset.loc[:, 'House Style'], drop_first=True)
Style2
Style3 = pd.get_dummies(dataset.loc[:, 'Garage'], drop_first=True)
Style3
```

```python
dataset = pd.concat([dataset, Style1], axis=1)
dataset = pd.concat([dataset, Style2], axis=1)
dataset = pd.concat([dataset, Style3], axis=1)
dataset = dataset.drop(['Neighborhood', 'House Style', 'Garage'], axis=1)
dataset
```

```python
X = dataset.loc[dataset['SalePrice'].notna(), :].drop('SalePrice', axis=1)
Y = dataset.loc[dataset['SalePrice'].notna(), 'SalePrice']

X['Total Bsmt SF'] = X['Total Bsmt SF'].fillna(X['Total Bsmt SF'].median())
```

```python
X_train, X_test, Y_train,  Y_test = train_test_split(
    X,
    Y,
    test_size=0.15,
    random_state=42

)
```

```python
model = LinearRegression()
model = model.fit(X_train, Y_train)
model.score(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
mean_absolute_error(Y_test, Y_pred)
```

```python
mean_squared_error(Y_test, Y_pred)
```
