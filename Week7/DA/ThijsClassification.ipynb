```python
df2 = pd.read_excel("AmesHousing.xlsx")
df2
```

```python
df2 = df2.loc[:, ("Gr Liv Area", "Year Built", "SalePrice", "Neighborhood", "House Style", 'Bedroom AbvGr', 'Overall Qual')]
df2
```

```python
dummies_df = pd.get_dummies(df2.loc[:, ("Neighborhood", "House Style")])
dummies_df
```

```python
df2 = pd.concat([df2, dummies_df], axis=1)
df2 = df2.drop(["House Style"], axis=1)
df2 = df2.drop(["Neighborhood"], axis=1)

df2
```

```python
X = df2.drop("Overall Qual", axis=1)
Y = df2.loc[:, "Overall Qual"]
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.15,
    random_state=42
)
```

```python
dtree = DecisionTreeClassifier(max_depth=10)
dtree = dtree.fit(X_train, Y_train)
tree.plot_tree(dtree, feature_names=X.columns)
plt.show()
```

```python
predicted_df2 = pd.DataFrame(dtree.predict(X_test))
predicted_df2 = predicted_df2.rename(columns= {0 : "Prediction"})
model_results_frame = pd.concat([Y_test.reset_index()['Overall Qual'], predicted_df2], axis=1)
model_results_frame
```

```python
labels = sorted(set(model_results_frame["Overall Qual"]))

confusionmatrix = metrics.confusion_matrix(model_results_frame["Overall Qual"], model_results_frame["Prediction"], labels=labels)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix= confusionmatrix, display_labels= labels)

cm_display.plot()
plt.show()
```

```python
metrics.accuracy_score(model_results_frame["Overall Qual"], model_results_frame["Prediction"])
```

```python
metrics.recall_score(model_results_frame["Overall Qual"], model_results_frame["Prediction"], average='weighted', zero_division=0)
```

```python
metrics.precision_score(model_results_frame["Overall Qual"], model_results_frame["Prediction"], average='weighted', zero_division=0)
```
