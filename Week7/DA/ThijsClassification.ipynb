```python
%pip install scikit-learn

import pandas as pd

from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics

import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("AmesHousing.xlsx")
df
```

```python
df = df.loc[:, ("Overall Qual", 'Garage')]
df
```

```python
# df = pd.concat([df, dummies_df], axis=1)
#  df = df.drop(["House Style"], axis=1)
#  df = df.drop(["Neighborhood"], axis=1)

# df
```

```python
X = df.drop("Garage", axis=1)
Y = df.loc[:, "Garage"]
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.15,
    random_state=42
)
```

```python
list(X.columns)
```

```python
dtree = DecisionTreeClassifier(max_depth=7)
dtree = dtree.fit(X_train, Y_train)
tree.plot_tree(dtree, feature_names=X.columns)
plt.show()
```

```python
predicted_df = pd.DataFrame(dtree.predict(X_test))
predicted_df = predicted_df.rename(columns= {0 : "Prediction"})
model_results_frame = pd.concat([Y_test.reset_index()['Garage'], predicted_df], axis=1)
model_results_frame
```

```python
confusionmatrix = metrics.confusion_matrix(model_results_frame["Garage"], model_results_frame["Prediction"])
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix= confusionmatrix, display_labels= [False, True])

cm_display.plot()
plt.show()
```

```python
metrics.accuracy_score(model_results_frame["Garage"], model_results_frame["Prediction"])
```

```python
metrics.recall_score(model_results_frame["Garage"], model_results_frame["Prediction"], pos_label='yes')
```

```python
metrics.precision_score(model_results_frame["Garage"], model_results_frame["Prediction"], pos_label='yes')
```
