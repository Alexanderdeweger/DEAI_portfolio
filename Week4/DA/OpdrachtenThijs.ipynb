1 Maak een eigen implementatie van een ’Singly linked list’. Er zijn meerdere implementaties denkbaar, maar we
adviseren om de list als abstracte class te implementeren met twee concrete subclasses. Een voor lege en een voor
niet-lege lijsten.

```python
from abc import ABC, abstractmethod
```

```python
class LinkedList(ABC):
    @staticmethod
    def makeList(numbers):
        result = LinkedListEmpty()
        for n in reversed(numbers):
            result = result.addFirst(n)
        return result

    @abstractmethod
    def isEmpty(self):
        pass

    @abstractmethod
    def toString(self):
        pass

    @abstractmethod
    def addFirst(self, value):
        pass

    @abstractmethod
    def remove(self, value):
        pass

    @abstractmethod
    def smallest(self):
        pass

    @abstractmethod
    def sortSimple(self):
        pass

    @abstractmethod
    def uniq(self):
        pass

    @abstractmethod
    def subList(self, start, end):
        pass

    @abstractmethod
    def merge(self, list):
        pass

    @abstractmethod
    def sortMerge(self):
        pass

    @abstractmethod
    def length(self):
        pass
```

```python
class LinkedListPopulated(LinkedList):

    def __init__(self, value, rest):
        self.value = value
        self.rest = rest

    def isEmpty(self):
        return False
    
    def toString(self):
        return str(self.value) + " " + self.rest.toString()
    
    def addFirst(self, value):
        return LinkedListPopulated(value, self)
    
    def remove(self, value):
        if self.value == value:
            return self.rest
        else:
            return LinkedListPopulated(self.value, self.rest.remove(value))

    def smallest(self):
        if self.rest.isEmpty():
            return self.value
        else:
            s = self.rest.smallest()
            return self.value if self.value < s else s

    def sortSimple(self):
        s = self.smallest()
        rest_sorted = self.remove(s).sortSimple()
        return rest_sorted.addFirst(s)

    def uniq(self):
        if self.value == self.rest.value:
            return self.rest.uniq()
        else:
            return 1 + self.rest.uniq()

    def subList(self, start, end):
        if end == 0:
            return LinkedListEmpty()
        if start == 0:
            return LinkedListPopulated(self.value, self.rest.subList(0, end - 1))
        return self.rest.subList(start - 1, end - 1)
        
    def merge(self, list):
        if list.isEmpty():
            return self
        if self.value <= list.value:
            return LinkedListPopulated(self.value, self.rest.merge(list))
        else:
            return LinkedListPopulated(list.value, self.merge(list.rest))
        
    def length(self):
        return 1 + self.rest.length()

    def sortMerge(self):
        n = self.length()
        if n == 1:
            return self
        left = self.subList(0, n // 2)
        right = self.subList(n // 2, n)
        return left.sortMerge().merge(right.sortMerge())
```

```python
class LinkedListEmpty(LinkedList):

    def isEmpty(self):
        return True

    def toString(self):
        return ""

    def addFirst(self, value):
        return LinkedListPopulated(value, self)

    def remove(self, value):
        return self

    def smallest(self):
        raise Exception("Lege lijst")

    def sortSimple(self):
        return self

    def uniq(self):
        return 0
    
    def subList(self, start, end):
        return self
    
    def merge(self, list):
        return list
    
    def length(self):
        return 0

    def sortMerge(self):
        return self
```
