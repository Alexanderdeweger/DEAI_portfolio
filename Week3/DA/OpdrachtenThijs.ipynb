1 Maak een eigen implementatie van een ’Singly linked list’. Er zijn meerdere implementaties denkbaar, maar we
adviseren om de list als abstracte class te implementeren met twee concrete subclasses. Een voor lege en een voor
niet-lege lijsten.

```python
from abc import ABC, abstractmethod
```

```python
class LinkedList(ABC):

    @abstractmethod
    def isEmpty(self):
        pass

    @abstractmethod
    def toString(self):
        pass

    @abstractmethod
    def addFirst(self, value):
        pass

    @abstractmethod
    def remove(self, value):
        pass

    @abstractmethod
    def smallest(self):
        pass

    @abstractmethod
    def sortSimple(self):
        pass

    @abstractmethod
    def uniq(self):
        pass
```

```python
class LinkedListPopulated(LinkedList):

    def __init__(self, value, rest):
        self.value = value
        self.rest = rest

    def isEmpty(self):
        return False
    
    def toString(self):
        return str(self.value) + " " + self.rest.toString()
    
    def addFirst(self, value):
        return LinkedListPopulated(value, self)
    
    def remove(self, value):
        if self.value == value:
            return self.rest
        else:
            return LinkedListPopulated(self.value, self.rest.remove(value))

    def smallest(self):
        if self.rest.isEmpty():
            return self.value
        else:
            s = self.rest.smallest()
            return self.value if self.value < s else s

    def sortSimple(self):
        s = self.smallest()
        rest_sorted = self.remove(s).sortSimple()
        return rest_sorted.addFirst(s)

    def uniq(self):
        if self.rest.isEmpty():
            return 1

        if self.value == self.rest.value:
            return self.rest.uniq()
        else:
            return 1 + self.rest.uniq()
```

```python
class LinkedListEmpty(LinkedList):

    def isEmpty(self):
        return True

    def toString(self):
        return ""

    def addFirst(self, value):
        return LinkedListPopulated(value, self)

    def remove(self, value):
        return self

    def smallest(self):
        raise Exception("Lege lijst")

    def sortSimple(self):
        return self

    def uniq(self):
        return self
```

7 Download de bestanden ’Read.py’ en ’kentekens1000.txt’ van Brightspace. Gebruik de code in ’Read.py’ om het

```python
import csv
import sys

sys.setrecursionlimit(20000000)

kentekens = LinkedListEmpty()

with open(r"kentekens1000.txt") as f:
    reader = csv.reader(f, delimiter=',')
    for row in reader:
        kentekens = kentekens.addFirst(row[0])

sorted_kentekens = kentekens.sortSimple()
print(sorted_kentekens.toString())
```
